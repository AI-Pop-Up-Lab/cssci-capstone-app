```python
# Recoded data frame

df_recoded = pd.DataFrame()

df_recoded['respondent_id'] = df['lopnr']

df_recoded['age_group'] = df['v7202'].astype(str).replace('nan', np.nan).map(age_7cut)

# constituency (register data) 
df_recoded['constituency_code'] = df['v7090']

# turnout (register validated)
df_recoded['turnout_2022'] = df['v7020'].map({0: 'Did not vote', 1: 'Voted'})

# main recode map
for dict_name, (src_col, new_col) in recode_map.items():
    if src_col in df.columns:
        mapping = recode_dicts[dict_name]
        src = df[src_col].astype(str).replace('nan', np.nan)  # convert categorical to string first
        mapped = src.map(mapping)
        df_recoded[new_col] = mapped.where(mapped.notna(), src)  # fall back to original if no match
    else:
        print(f"Warning: column {src_col} not found in dataframe")

for src_col, new_col in agree_disagree_items.items():
    if src_col in df.columns:
        src = df[src_col].astype(str).replace('nan', np.nan)
        mapped = src.map(agree_disagree)
        df_recoded[new_col] = mapped.where(mapped.notna(), src)

for src_col, new_col in policy_items.items():
    if src_col in df.columns:
        src = df[src_col].astype(str).replace('nan', np.nan)
        mapped = src.map(good_bad)
        df_recoded[new_col] = mapped.where(mapped.notna(), src)

for src_col, new_col in trust_items.items():
    if src_col in df.columns:
        src = df[src_col].astype(str).replace('nan', np.nan)
        mapped = src.map(trust)
        df_recoded[new_col] = mapped.where(mapped.notna(), src)


numeric_renames = {
    'q307':  'democracy_scale_sweden',       # 0-10
    'q324':  'vote_makes_difference',        # 1-5
    'q328':  'self_leftright_placement',     # 0-10
    'q336':  'election_fairness',            # 1-5
    **party_likedislike_items,               # -5 to +5
    **party_leftright_items,                 # 0-10
    **society_aims_items,                    # 0-10
    # news consumption days per week
    'q302a': 'days_news_svt',
    'q302b': 'days_news_tv4',
    'q302c': 'days_news_radio',
    'q302d': 'days_newspapers',
    'q302e': 'days_online_news',
    'q302f': 'days_social_media',
}

for src_col, new_col in numeric_renames.items():
    if src_col in df.columns:
        df_recoded[new_col] = pd.to_numeric(df[src_col], errors='coerce')

# crisis leader gender (two separate questions)
if 'q344' in df.columns:
    df_recoded['better_economic_crisis_leader'] = df['q344'].map(crisis_leader_gender)
if 'q345' in df.columns:
    df_recoded['better_health_crisis_leader'] = df['q345'].map(crisis_leader_gender)

# non-voters in past_vote
df_recoded.loc[df_recoded['turnout_2022'] == 'Did not vote', 'past_vote'] = 'Did not vote'

print(f"Shape: {df_recoded.shape}")
print(f"\nVote in 2022:")
print(df_recoded['past_vote'].value_counts(dropna=False))
print(f"\nGender:")
print(df_recoded['gender'].value_counts(dropna=False))
print(f"\nEducation:")
print(df_recoded['education_level'].value_counts(dropna=False))
print(f"\nAge group:")
print(df_recoded['age_group'].value_counts(dropna=False).sort_index())
```
