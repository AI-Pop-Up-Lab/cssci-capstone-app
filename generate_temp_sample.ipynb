# Temporary sample generation notebook

In the future, the idea is have the samples to be generated daily, automatically and dynamically

for the pilot test, the sample is pregenerated

```python
import pandas as pd
```

```python
# read the CSV into a DataFrame
df = pd.read_csv(
    './ai-popup-lab-backend/country_data/stratification_frames/sweden_strat_frame.csv'
)


df.head()
```

```python
# Sample 10,000 people from the dataframe, where each row can be sampled N times

df_sampled = df.loc[df.index.repeat(df['N'].astype(int))].sample(n=10000, random_state=42).reset_index(drop=True)

len(df_sampled)
```

```python
df_sampled = df_sampled.drop('N', axis = 1)
df_sampled = df_sampled.drop('MunicipalityCode', axis=1)
df_sampled.head()
```

```python
df_sampled.reset_index().to_csv('sweden_daily_sample.csv', index=False)
```
