# Unique column value generation tool

this notebook generates a data structure for the country_data_info.json in ai-popup-lab-backend/country_data

this is then fed through the backend to the frontend to give the options for dropdowns when the user wants to choose demographics to see more specific demographic tables (or for choosing the persona)

```python
import pandas as pd

df = pd.read_csv(r'ai-popup-lab-backend\country_data\surveys\usa_survey.csv') #replace with daily sample of country you wish to generate for (rather than strat frame where values may not be translated or edited correctly)
df.head()
```

```python
# place here the relevant columns for the demographic choices in the front-end
relevant_columns = ["age_cat", "gender_cat", "race_cat", "hispanic_cat", "educ_cat", "state_abbrv", "vote_choice"]


# create dictionary to hold unique values for each relevant column
unique_values = {col: df[col].dropna().unique().tolist()
                 for col in relevant_columns}
```

# output

now copy and past the resulting dictionary printed below into the *country_data_info.json* file, in the correct country, value of key "column_unique_vals"

```python
import json

print(json.dumps(unique_values))
```
